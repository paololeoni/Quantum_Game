```python
import qiskit
from qiskit import transpile
from qiskit import QuantumCircuit
from qiskit_aer import AerSimulator
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt
import numpy as np

def simulate(quantum_circuit=None):
    ## Initialize the simulator
    sim = AerSimulator()

    ## Generates a transpiled version of the quantum circuit that is optimized for the specified backend (simulator in this case)
    tqc = transpile(quantum_circuit, sim)
    ## Run the simulation with 1000 shots (repetitions) to get a distribution of results
    result = sim.run(tqc, shots=1000).result()
    return result
```

## Superposition

```python
## Create a quantum circuit with one qubit
qc_superposition = QuantumCircuit(1,1)
# Apply a certain gate taught before and measure the qubit
qc_superposition.h(0)
qc_superposition.measure(0,0)
# Use the simulate function
result = simulate(qc_superposition)
# Plot the result and draw the circuit
plot_histogram(result.get_counts())
```

## Introduction Bias


```python
## Create a quantum circuit with one qubit
qc_bias = QuantumCircuit(1,1)
# Apply a certain gate taught before and measure the qubit
qc_bias.h(0)
qc_bias.ry(-np.pi/6,0)
qc_bias.measure(0,0)
# Use the simulate function
result = simulate(qc_bias)
# Plot the result and draw the circuit
plot_histogram(result.get_counts())
```

## Flipping the coin

```python
## Create a quantum circuit with one qubit
qc_flip = QuantumCircuit(1,1)
# Apply a certain gate taught before and measure the qubit
qc_flip.h(0)
qc_flip.ry(-np.pi/6,0)
qc_flip.x(0)
qc_flip.measure(0,0)
# Use the simulate function
result = simulate(qc_flip)
# Plot the result and draw the circuit
plot_histogram(result.get_counts())
```

## Creating Entanglement

```python
# Create a quantum circuit with two qubits
qc_two = QuantumCircuit(2, 2)
# Apply Hadamard gate to qubit 0, notice we apply it to only one qubit, because we also need a control qubit
qc_two.h(0)
qc_two.cx(0, 1)
qc_two.measure([0,1], [0,1])  # measure both qubits we need to store both qubits into two list

# Simulate the result
result = simulate(qc_two)

plot_histogram(result.get_counts())
```
